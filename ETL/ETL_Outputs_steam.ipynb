```python
#importamos las librerias
import pandas as pd
import json
import warnings
warnings.filterwarnings("ignore")
```

```python
#Se extraen los datos desde el archivo 'steam_games.json', se convierte en Dataframe y se guarda como csv

with open('D:\Desktop\proyecto entrega 1\json\output_steam_games.json', 'r', encoding='utf-8') as archivo:
   steam = [json.loads(line) for line in archivo]

#Convertirlo a DataFrame
out_steam = pd.DataFrame(steam)

#Guardar el Dataframe como csv
out_steam.to_csv("steam.csv", index=False)

print("OK")
```

```python
#Visualización del DataFrame
out_steam
```

```python
#Analisis de los valores nulos y los tipos de datos

def info_nulos(dataframe):
    datos = {"Columna": [], "tipo_de_dato": [], "%_no_nulos": [], "%_nulos": [], "nulos": []}

    for columna in dataframe.columns:
        no_nulos = (dataframe[columna].count() / len(dataframe)) * 100
        datos["Columna"].append(columna)
        datos["tipo_de_dato"].append(dataframe[columna].apply(type).unique())
        datos["%_no_nulos"].append(round(no_nulos, 2))
        datos["%_nulos"].append(round(100-no_nulos, 2))
        datos["nulos"].append(dataframe[columna].isnull().sum())

    df_info = pd.DataFrame(datos)
        
    return df_info

info_nulos(out_steam)
```

```python
#eliminamos los archivos NaN
out_steam= out_steam.dropna(how='all').reset_index(drop=True)
out_steam.shape
```

```python
#Se procede a completar los valores nulos y convertir a tipo float aquellos que no lo son, previamente asignando el valor 'free to play' a 0.
out_steam['price'] = pd.to_numeric(out_steam['price'], errors='coerce') #pasamos a numerica la columna
out_steam['price'] = out_steam['price'].replace('Free To Play', 0) #se reemplaza por 0 el free to play
out_steam['price'] = out_steam['price'].fillna(0)
```
