```python
#Verificamos los tipos de datos y los nulos
def info_nulos(dataframe):
    datos = {"Columna": [], "tipo_de_dato": [], "%_no_nulos": [], "%_nulos": [], "nulos": []}

    for columna in dataframe.columns:
        no_nulos = (dataframe[columna].count() / len(dataframe)) * 100
        datos["Columna"].append(columna)
        datos["tipo_de_dato"].append(dataframe[columna].apply(type).unique())
        datos["%_no_nulos"].append(round(no_nulos, 2))
        datos["%_nulos"].append(round(100-no_nulos, 2))
        datos["nulos"].append(dataframe[columna].isnull().sum())

    df_info = pd.DataFrame(datos)
        
    return df_info

info_nulos(user_review)
```

```python
#Se crea la columna 'sentiment_analysis' 

def analisis_sentimientos(review):
  sentimiento = analyzer.polarity_scores(review)
  if sentimiento['compound'] <-0.5:
      return 0 # Malo
  elif sentimiento['compound'] > 0.5:
      return 2 # Positivo
  else:
      return 1 # Neutral


user_review['sentiment_analysis'] = user_review['review'].apply(analisis_sentimientos)

#Reemplazar la columna 'review' con la columna 'sentiment_analysis'
user_review['review'] = user_review['sentiment_analysis'] #
# Rellena los datos nulos con 1
user_review['review'] = user_review['review'].fillna(1) 
#elimina la columna review
user_review = user_review.drop('review', axis=1) 

```

```python
#Primero se debe manejar el formato de la fecha debido a que vienen de diferentes formas
def transform_date(date_str):
   # Dividir la cadena por espacios
   parts = date_str.split()

   # Obtener el día y el mes
   day = parts[2].replace(',', '')
   month = parts[1]

   # Mapear el nombre del mes a su número correspondiente
   month_dict = {
       'January': '01', 'February': '02', 'March': '03', 'April': '04', 'May': '05', 'June': '06',
       'July': '07', 'August': '08', 'September': '09', 'October': '10', 'November': '11', 'December': '12'
   }
   if month in month_dict:
       month = month_dict[month]

   # Obtener el año si está presente, si no, usar 1900
   year = parts[-1] if len(parts) == 4 else '1900'

   # Formatear la fecha
   formatted_date = f"{year}-{month}-{day}"
   
   return formatted_date

user_review['posted'] = user_review['posted'].apply(transform_date)
```
