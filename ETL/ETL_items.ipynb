```python
#Importamos librerias
import pandas as pd
import json
import ast
import warnings
warnings.filterwarnings("ignore")
```

```python
#Se extraen los datos desde el archivo 'users_items.json', se convierte en Dataframe y se guarda como csv

#Cargar el archivo
ruta_items = r'D:\Desktop\proyecto entrega 1\json\\australian_users_items.json'


# Se lee de cada línea del dataset
filas_items = []
with open(ruta_items, 'r', encoding='utf-8') as f:
   for line in f.readlines():
       filas_items.append(ast.literal_eval(line))

# Se convierte en dataframe
user_items = pd.DataFrame(filas_items)

user_items.to_csv("user_items.csv", index=False)

print("OK")
```

```python
#Se verifican los tipos de datos para cada columna y la cantidad de nulos
def info_nulos(dataframe):
    datos = {"Columna": [], "tipo_de_dato": [], "%_no_nulos": [], "%_nulos": [], "nulos": []}

    for columna in dataframe.columns:
        no_nulos = (dataframe[columna].count() / len(dataframe)) * 100
        datos["Columna"].append(columna)
        datos["tipo_de_dato"].append(dataframe[columna].apply(type).unique())
        datos["%_no_nulos"].append(round(no_nulos, 2))
        datos["%_nulos"].append(round(100-no_nulos, 2))
        datos["nulos"].append(dataframe[columna].isnull().sum())

    df_info = pd.DataFrame(datos)
        
    return df_info

info_nulos(user_items)
```

```python
#Se normalizan los datos 
user_items = pd.json_normalize(filas_items, record_path=['items'], meta=['steam_id','items_count','user_id', 'user_url'] )  
user_items
```

Analizamos la columna 'user_id' que a pesar que debe ser unica, se encuentra repetida porque un solo usuario puede jugar variado pero se compara la duplicidad en relacion con "item_id", "item_name" y "steam_id"

```python
df_items_repetidos_id = user_items[['user_id', 'item_id', 'item_name', 'steam_id']].duplicated().sum() #Cantidad de valores repetidos 
print("El numero de 'user_id' repetidos en relación con 'item_id', 'item_name' y 'steam_id' es de:", df_items_repetidos_id)
```

```python
#df que contiene solo las filas duplicadas (excepto la primera ocurrencia) 
df_items_repetidos_id  = user_items[['user_id', 'item_id', 'item_name', 'steam_id']][user_items[['user_id', 'item_id','item_name','steam_id']].duplicated(keep=False)] 
df_items_repetidos_id
```
